Here I am Merging various datasets so that I am getting a wide knowledge about covid vaccination distribution and which would be a great example and follow-up for anyother pandemic

```python
from datetime import date
import pandas as pd
# This is the URLs of your CSV files
url1 = 'https://data.cdc.gov/api/views/km4m-vcsb/rows.csv?accessType=DOWNLOAD'
url2 = 'https://data.cdc.gov/api/views/unsk-b7fc/rows.csv?accessType=DOWNLOAD'
# here it reads CSV files into DataFrames
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
print("Columns in df1:", df1.columns)
print("Columns in df2:", df2.columns)
common_column_name = 'Date'
# Here I am trying to check if both the dataset have a common column
if common_column_name in df1.columns and common_column_name in df2.columns:
    # Merge DataFrames
    merged_df = pd.merge(df1, df2, on=common_column_name, how='inner')

    print(merged_df)
    print("successfully merged")
else:
    print(f"Common column '{date}' not found in both DataFrames.")
```

Adding another set of dataset so that I wanted even more wide variety of knowledge about covid vaccine distribution

```python
import pandas as pd
url1 = 'https://data.cdc.gov/api/views/km4m-vcsb/rows.csv?accessType=DOWNLOAD'
url2 = 'https://data.cdc.gov/api/views/unsk-b7fc/rows.csv?accessType=DOWNLOAD'
url3 = 'https://data.cdc.gov/api/views/hwk8-wu83/rows.csv?accessType=DOWNLOAD'
# Here I am trying to read CSV files into DataFrames
df1 = pd.read_csv(url1)
df2 = pd.read_csv(url2)
df3 = pd.read_csv(url3)
print("Columns in df1:", df1.columns)
print("Columns in df2:", df2.columns)
print("Columns in df3:", df3.columns)
common_column_name1 = 'Date'
common_column_name2 = 'Date'

if (
    common_column_name1 in df1.columns
    and common_column_name1 in df2.columns
    and common_column_name1 in df3.columns
    and common_column_name2 in df1.columns
    and common_column_name2 in df2.columns
    and common_column_name2 in df3.columns
):

    merged_df = pd.merge(df1, df2, on=[common_column_name1, common_column_name2], how='inner')
    merged_df = pd.merge(merged_df, df3, on=[common_column_name1, common_column_name2], how='inner')

    print(merged_df)
    print("Successfully merged.")
else:
    print("One or more common columns not found in all three DataFrames.")
```

Aggregating certain columns so that I wanted distribution information together rather than putting it seperately

```python
import pandas as pd
url = 'https://data.cdc.gov/api/views/unsk-b7fc/rows.csv?accessType=DOWNLOAD'
df = pd.read_csv(url)
print("Original DataFrame:")
print(df.head())
# Here I am trying to aggregate
columns_to_aggregate = ['Distributed', 'Distributed_Janssen', 'Distributed_Moderna', 'Distributed_Pfizer', 'Distributed_Novavax', 'Distributed_Unk_Manuf']
aggregated_df = df.groupby('Date')[columns_to_aggregate].sum().reset_index()
print("\nAggregated DataFrame:")
print(aggregated_df)
```

Here I am doing the sumation of different dataset because its different data of distribution so that in future it will help healthcare to know how much vaccine should be distribute for a particular region and the would be prepared enough

```python
import pandas as pd
url = 'https://data.cdc.gov/api/views/hwk8-wu83/rows.csv?accessType=DOWNLOAD'
df = pd.read_csv(url)
print("Original DataFrame:")
print(df.head())
columns_to_aggregate = ['Distributed_Bivalent', 'Dist_Bivalent_MOD', 'Dist_Bivalent_PFR', 'Distributed_Novavax']
aggregated_df = df.groupby('Date')[columns_to_aggregate].sum().reset_index()
print("\nAggregated DataFrame:")
print(aggregated_df)
```

Creating a new column of named Total for doing some Data Transformation

```python
import pandas as pd
url = 'https://data.cdc.gov/api/views/km4m-vcsb/rows.csv?accessType=DOWNLOAD'
df = pd.read_csv(url)
print("Original DataFrame:")
print(df.head())
# Here I am trying to add all the columns thats listed below
columns_to_add = ['Administered_Dose1', 'Administered_Dose1_pct_known', 'Administered_Dose1_pct_US', 'Series_Complete_Yes', 'Administered_Dose1_pct_agegroup']
# here I am creating a new column 'Total'
df['Total'] = df[columns_to_add].sum(axis=1)
print("\nDataFrame with Total column:")
print(df.head())
```

In this I am trying to perform some data transformations by doing duplicate of a column where I wanted some backup of a important column

```python
import pandas as pd
url = 'https://data.cdc.gov/api/views/hwk8-wu83/rows.csv?accessType=DOWNLOAD'
df = pd.read_csv(url)
print("Original DataFrame:")
print(df.head())
# This is the column that needs to be duplicate
column_to_duplicate = 'Distributed_Novavax'
df['Duplicated_Column'] = df[column_to_duplicate]
print("\nDataFrame with Duplicated_Column:")
current_column_name = 'Duplicated_Column'
new_column_name = 'Distributed_Duplicate'
df.rename(columns={current_column_name: new_column_name}, inplace=True)
print("\nDataFrame with Renamed Column:")
print(df.head())
```

Dropping that duplicated column

```python
column_to_delete = 'Distributed_Duplicate'
df.drop(columns=column_to_delete, inplace=True)
print("\nDataFrame after deleting column:")
print(df.head())
```

This Visualisation that I made gives a clear visualization of people who took Dose 1 without hesitation and also it gives at which date the people got more number of dose 1 so that it gives a clear idea of how much days it is taking for a person to accept a new medicine

```python
import pandas as pd
import matplotlib.pyplot as plt
url = 'https://data.cdc.gov/api/views/km4m-vcsb/rows.csv?accessType=DOWNLOAD'
df = pd.read_csv(url)
print("Original DataFrame:")
print(df.head())
x_column = 'Date'
y_column = 'Administered_Dose1'
plt.plot(df[x_column], df[y_column])
plt.xlabel('Date')
plt.ylabel('Administered Dose 1')
plt.title('Administered Dose 1 Over Time')
plt.show()
```

This graph gives a clear picture of distribution of vaccines over the different locations and in future we will have a clear picture of how many vaccines needed for a particular location.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
url = 'https://data.cdc.gov/api/views/unsk-b7fc/rows.csv?accessType=DOWNLOAD'
df = pd.read_csv(url)
print("Original DataFrame:")
print(df.head())
x_column = 'Location'
y_column = 'Distributed'
plt.figure(figsize=(12, 6))
sns.barplot(data=df, x=x_column, y=y_column)
plt.xlabel('Location')
plt.ylabel('Distributed')
plt.title('Distribution Across Locations')
plt.xticks(rotation=45, ha='right')
plt.show()
```
